```python
#Importing country Data
popu = pd.read_csv('population_by_country_2020.csv')
popu = popu.rename(columns = {'Country (or dependency)':'country', 'Population (2020)':'population',
                                         'Med. Age':'median_age', 'Urban Pop %':'urban_percentage_share',
                                         'World Share':'world_share'})
popu = popu.drop(['Yearly Change', 'Net Change', 'Density (P/Km²)', 'Land Area (Km²)', 'Migrants (net)',
                             'Fert. Rate'], axis =1)
popu
```

```python
##Function to get diff country 
def similar_and_non_similar_Countries(main, array1, array2):
    similar = []
    not_in_1 = []
    not_in_2 = []
    totally_diff = []
    for i in main:
        if (i in array1) and (i in array2):
            similar.append(i)
        elif (i in array1) and (i not in array2):
            not_in_2.append(i)
        elif (i not in array1) and (i in array2):
            not_in_1.append(i)
        else:
            totally_diff.append(i)
    
    return similar, not_in_1, not_in_2, totally_diff
```

```python
dfcn1 = np.array(df.country.unique())
popucn2 = np.array(popu.country.unique())
datacn1 = np.array(data.country.unique())
```

```python
similar, not_in_1, not_in_2, totally_diff = similar_and_non_similar_Countries(dfcn1, datacn1, popucn2)
```

```python
def leftoff(check, same):
    output = []
    
    for i in range(0, len(check)):
        if check[i] not in same:
            output.append(check[i])
        else:
            continue;
    return output
```

```python
def dropoff(dfs, data_left):
    for i in range(0, len(data_left)):
        ele = data_left[i]
        idx = dfs[dfs.country == ele].index
        dfs.drop(idx, axis = 0, inplace = True)
    
    return dfs
```
